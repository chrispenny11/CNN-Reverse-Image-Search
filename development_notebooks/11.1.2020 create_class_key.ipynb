```python
import pandas as pd
import numpy as np
import os
from keras.applications.resnet import decode_predictions
```

```python
cwd = os.getcwd()
print(cwd)
```

```python
df_imagenet_key = pd.read_csv('./ImageNet/imagenet_key.csv')
# display(df_imagenet_key.transpose().reset_index())
df_imagenet_key = df_imagenet_key.transpose().reset_index()
# display(df_imagenet_key)

df_imagenet_key = df_imagenet_key.drop(labels = ['index', 4, 5, 6, 7], axis = 1)
# display(df_imagenet_key)

df_imagenet_key.columns = ['val_id', 'train_id', 'name', 'description']
display(df_imagenet_key)
```

```python
# print(df_imagenet_key.columns)
# display(df_imagenet_key[[1]])
# df_imagenet_key[df_imagenet_key[1].str.match('n02966687', na = False)]
```

```python
class_list = []
id_list = []
description_list = []

for i in range(1000):
    result_vector = np.zeros((1, 1000))
    result_vector[0, i] = 1
    decoded = decode_predictions(result_vector, top=1)
#     print(decoded[0][0])
    class_list.append(int(i) + 1)
    id_list.append(decoded[0][0][0])
    description_list.append(decoded[0][0][1])
    
    
df_resnet_key = pd.DataFrame({'resnet_id': id_list, 'resnet_desc': description_list, 'resnet_class': class_list})
# display(df_resnet_key)
```

```python
df_merged_keys = pd.merge(df_imagenet_key, df_resnet_key, how = 'inner', left_on = 'train_id', right_on = 'resnet_id')
display(df_merged_keys)
```

```python
df_merged_keys.to_csv('./ImageNet/imagenet_resnet_key.csv')
```
