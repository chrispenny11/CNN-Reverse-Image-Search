```python
def process_scores(score_src, ranks = [5, 10, 20, 50]):
    
    score_src = score_src
    
    with open(score_src) as f:
        loaded_scores =  json.load(f)
        
#     print(loaded_scores)
    
#     mean_scores_by_class = {}
#     for idx in range(1, 1001):
#         for rank in ranks:
#             mean_scores_by_class[str(idx)][str(rank)] = np.mean(loaded_scores[str(idx)][str(rank)])
    
#     print(mean_scores_by_class)
    def map_at_rank(rank):

        class_averages = []

        for (k, v) in loaded_scores.items():
#             class_averages.append(np.mean(v[str(rank)]))
            if k != "time":
#                 print(v)
                class_averages.append(np.mean(v[str(rank)]))
#     map_at_rank('1')
        return np.mean(class_averages)#, np.argsort(class_averages)[::-1]
    
    return [map_at_rank('5'), map_at_rank('10'), map_at_rank('20'), map_at_rank('50')]


print(process_scores("./annoy_1000_resnet_1000_scores.txt"))
print(process_scores("./annoy_1000_resnet_2048_scores.txt"))
print(process_scores("./annoy_1000_combined_scores.txt"))
print(process_scores("./annoy_1000_combined_0.4_scores.txt"))
print(process_scores("./annoy_1000_ae_bottleneck_2048_scores.txt"))
print(process_scores("./annoy_1000_ae_dense_2048_scores.txt"))
print(process_scores("./annoy_1000_ae_simple_2048_scores.txt"))

print(process_scores("./annoy_100_resnet_1000_scores.txt"))
print(process_scores("./annoy_100_resnet_2048_scores.txt"))
print(process_scores("./annoy_100_combined_scores.txt"))
print(process_scores("./annoy_100_combined_0.4_scores.txt"))
print(process_scores("./annoy_100_ae_bottleneck_2048_scores.txt"))
print(process_scores("./annoy_100_ae_dense_2048_scores.txt"))
print(process_scores("./annoy_100_ae_simple_2048_scores.txt"))

print(process_scores("./annoy_10_resnet_1000_scores.txt"))
print(process_scores("./annoy_10_resnet_2048_scores.txt"))
print(process_scores("./annoy_10_combined_scores.txt"))
print(process_scores("./annoy_10_combined_0.4_scores.txt"))
print(process_scores("./annoy_10_ae_bottleneck_2048_scores.txt"))
print(process_scores("./annoy_10_ae_dense_2048_scores.txt"))
print(process_scores("./annoy_10_ae_simple_2048_scores.txt"))

print(process_scores("./brute_force_resnet_1000_scores.txt"))
print(process_scores("./brute_force_normalized_resnet_1000_scores.txt"))
print(process_scores("./annoy_10000_resnet_1000_scores.txt"))
print(process_scores("./annoy_1000_resnet_1000_normalized_scores.txt"))
```

```python
def plot_histograms(score_src, rank = 20):
    
    score_src = score_src
    
    with open(score_src) as f:
        loaded_scores =  json.load(f)
        
#     print(loaded_scores)
    
#     mean_scores_by_class = {}
#     for idx in range(1, 1001):
#         for rank in ranks:
#             mean_scores_by_class[str(idx)][str(rank)] = np.mean(loaded_scores[str(idx)][str(rank)])
    
#     print(mean_scores_by_class)
    def make_histogram():

        class_averages = []

        for (k, v) in loaded_scores.items():
#             class_averages.append(np.mean(v[str(rank)]))
            if k != "time":
#                 print(v)
                class_averages.append(np.mean(v[str(rank)]))
            
#         plt.clf()
        plt.hist(class_averages, bins = 100, alpha = 0.5)
        plt.show()
        
        return class_averages
    
    class_averages = make_histogram()    
        
    return class_averages

ca1 = plot_histograms("./annoy_1000_resnet_1000_scores.txt")
ca2 = plot_histograms("./annoy_1000_combined_0.4_scores.txt")
ca3 = plot_histograms("./annoy_1000_ae_bottleneck_2048_scores.txt")
ca4 = plot_histograms("./annoy_1000_ae_dense_2048_scores.txt")
ca5 = plot_histograms("./annoy_1000_ae_simple_2048_scores.txt")

plt.clf()
plt.hist(ca1, bins = 50, alpha = 0.45, label = 'ResNet 1,000')
plt.hist(ca2, bins = 50, alpha = 0.45, label = 'Combined Weighted 3,048')
plt.legend()
plt.title("Distribution of Average Precision at Rank 20 by Class, Bins = 50")
plt.xlabel("Precision at Rank 20")
plt.ylabel("Frequency Count")
plt.show()

plt.clf()
plt.hist(ca3, bins = 100, alpha = 0.45, label = 'AE Bottleneck')
plt.hist(ca4, bins = 100, alpha = 0.45, label = 'AE Dense')
plt.hist(ca5, bins = 100, alpha = 0.45, label = 'AE Simple')
plt.legend()
plt.title("Distribution of Average Precision at Rank 20 by Class, Bins = 100")
plt.xlabel("Precision at Rank 20")
plt.ylabel("Frequency Count")
plt.show()
```
