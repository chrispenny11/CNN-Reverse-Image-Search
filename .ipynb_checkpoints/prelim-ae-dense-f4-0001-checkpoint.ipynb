```python
import tensorflow
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.applications as ka
import numpy as np
import sys
import os
# import boto3
# from sagemaker import get_execution_role
from PIL import Image
# import urllib.request

# role = get_execution_role()

import matplotlib
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]=3,3
```

```python
# Load resnet50 as base model:
base_model = ka.ResNet50(
                        weights='imagenet',
                        input_shape = (224, 224, 3)
                        )

# Handoff:
handoff_layer = base_model.layers[-3].output

# # Fully connected pathway:
ae_gmp = keras.layers.GlobalMaxPool2D()(handoff_layer)
ae_dense = keras.layers.Dense(12544, activation='relu')(ae_gmp)
ae_reshape = keras.layers.Reshape((112, 112, 1))(ae_dense)

# Append on decoding layers as convolutions followed by 2D upsampling:
ae_conv1 = keras.layers.Conv2D(512, (4, 4), padding = 'same', activation='relu')(handoff_layer)
ae_up_1 = keras.layers.UpSampling2D((2, 2))(ae_conv1)
ae_conv2 = keras.layers.Conv2D(256, (4, 4), padding = 'same', activation='relu')(ae_up_1)
ae_up_2 = keras.layers.UpSampling2D((2, 2))(ae_conv2)
ae_conv3 = keras.layers.Conv2D(128, (4, 4), padding = 'same', activation='relu')(ae_up_2)
ae_up_3 = keras.layers.UpSampling2D((2, 2))(ae_conv3)
ae_conv4 = keras.layers.Conv2D(56, (4, 4), padding = 'same', activation='relu')(ae_up_3)
ae_up_4 = keras.layers.UpSampling2D((2, 2))(ae_conv4)

# Add dense connection as an additional channel:
ae_concat = keras.layers.Concatenate()([ae_up_4, ae_reshape])

ae_conv5 = keras.layers.Conv2D(28, (4, 4), padding = 'same', activation='relu')(ae_concat)
# ae_conv5 = keras.layers.Conv2D(28, (4, 4), padding = 'same', activation='relu')(ae_concat)
ae_up_5 = keras.layers.UpSampling2D((2, 2))(ae_conv5)
ae_conv6 = keras.layers.Conv2D(3, (4, 4), padding = 'same', activation='relu')(ae_up_5)

# Define the model that includes our decoding layers:
model = keras.Model(inputs = [base_model.input], outputs = ae_conv6)#, base_model.output, base_model.layers[-2].output])
# Define optimizer:
# opt = keras.optimizers.Adam(learning_rate = 0.0001, decay = 0.01)
opt = keras.optimizers.Adam(learning_rate = 0.0001)
# Compile model:
model.compile(optimizer = opt, loss = 'mse')

# Describe:
model.summary()
```

```python
train_directory = '/kaggle/input/imagenetb1/'
val_directory = '/kaggle/input/imagnetb2/'

train_IDG = keras.preprocessing.image.ImageDataGenerator()
val_IDG = keras.preprocessing.image.ImageDataGenerator()

train_gen = train_IDG.flow_from_directory(
    train_directory, 
    target_size=(224, 224),
    color_mode='rgb', 
    class_mode='input', 
    batch_size=64, 
    shuffle=True, 
    seed=13
)

val_gen = val_IDG.flow_from_directory(
    val_directory, 
    target_size=(224, 224),
    color_mode='rgb', 
    class_mode='input', 
    batch_size=128, 
    shuffle=True, 
    seed=13
)
```

```python
history1 = model.fit(train_gen, validation_data = val_gen, epochs = 100, shuffle = True, workers = 6, verbose = 1)
```

```python
train_loss = history1.history['loss']
val_loss = history1.history['val_loss']

plt.clf()
plt.plot([str(x) for x in range(1, 101)], train_loss, c = 'c', label = 'Train Loss')
plt.plot([str(x) for x in range(1, 101)], val_loss, c = 'm', label = 'Validation Loss')
plt.legend()
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
```
