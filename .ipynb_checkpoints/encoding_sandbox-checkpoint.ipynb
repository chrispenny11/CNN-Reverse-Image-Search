```python
import numpy as np
import sys
# import sklearn.model_selection as skl
# from sklearn.model_selection import train_test_split
import pickle

import matplotlib
import matplotlib.pyplot as plt
import time
import PIL

import os

from annoy import AnnoyIndex

plt.rcParams["figure.figsize"]=1,1
```

```python
# https://wiki.python.org/moin/UsingPickle

chunk_1 = pickle.load(open("imagenet_val_2048_encoding_" + str(1) + ".p", "rb") ) 
chunk_2 = pickle.load(open("imagenet_val_2048_encoding_" + str(2) + ".p", "rb") ) 
chunk_3 = pickle.load(open("imagenet_val_2048_encoding_" + str(3) + ".p", "rb") ) 
chunk_4 = pickle.load(open("imagenet_val_2048_encoding_" + str(4) + ".p", "rb") ) 
chunk_5 = pickle.load(open("imagenet_val_2048_encoding_" + str(5) + ".p", "rb") ) 

encodings = np.concatenate([chunk_1, chunk_2, chunk_3, chunk_4, chunk_5], axis = 0)
```

```python
print(encodings.shape)
```

```python
# Use Brute Force:

# %%timeit

start_time = time.time()

first = encodings[2, :]
dist_from_first = []


for idx, item in enumerate(encodings):
    dist_from_first.append((np.linalg.norm(first-item), idx))
        
dist_from_first.sort()

end_time = time.time()

print(end_time - start_time)

print(dist_from_first)

```

```python
# Use Annoy:

t = AnnoyIndex(2048, 'euclidean')

for idx, item in enumerate(encodings):
    t.add_item(idx, item)
    
# t.get_nns_by_item(0, 100)
t.build(100)
t.save('imagenet_val_encodings.ann')
```
