```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"]= 14,9
```

```python
df_comparison = pd.read_csv('./results/11.03.2020 baseline_resnet_benchmarks.csv')
display(df_comparison)

prec_compare = df_comparison[['Precision at 5', 'Precision at 10', 'Precision at 15', 'Precision at 20']].values
print(prec_compare)
```

```python
plt.clf()
plt.scatter(['5', '10', '15', '20'], prec_compare[0], color = '#6E33FF', label = 'Brute Force, 1000 Float Encoding', s = 90)
plt.scatter(['5', '10', '15', '20'], prec_compare[1], color = '#1D8BEC', label = 'Brute Force, 2048 Float Encoding', s = 90)
plt.legend()
plt.title('Precision at Rank: Brute Force Nearest Neighbor')
plt.ylabel('Precision at Rank')
plt.xlabel('Rank (Returned Image Results)')
plt.show()
```

```python
plt.clf()
plt.scatter(['5', '10', '15', '20'], prec_compare[0], color = '#6E33FF', label = 'Brute Force, 1000 Encoding', s = 90)
plt.scatter(['5', '10', '15', '20'], prec_compare[1], color = '#1D8BEC', label = 'Brute Force, 2048 Encoding', s = 90)
plt.scatter(['5', '10', '15', '20'], prec_compare[2], color = '#B61736', label = 'ANNOY, Single Tree, 1000 Encoding', s = 90)
plt.scatter(['5', '10', '15', '20'], prec_compare[3], color = '#D11A9A', label = 'ANNOY, Single, Tree, 2048 Encoding', s = 90)
plt.legend()
plt.title('Precision at Rank: Brute Force Nearest Neighbor')
plt.ylabel('Precision at Rank')
plt.xlabel('Rank (Returned Image Results)')
plt.show()
```
